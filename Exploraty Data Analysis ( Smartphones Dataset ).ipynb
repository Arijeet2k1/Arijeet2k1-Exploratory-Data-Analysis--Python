column --> PIXLES_IN_WIDTH and PIXLES_IN_HEIGHT

Feature Engineering on ( DISPLAY_IN_INCHES , PIXLES_IN_WIDTH , PIXLES_IN_HEIGHT ) columns to extract Pixels Per Inch so that Resolution of phones can be extracted

```python
# Calculating pixel density (PPI) for each row
smartphones['Pixel Density'] = (smartphones['PIXLES_IN_WIDTH']**2 + smartphones['PIXLES_IN_HEIGHT']**2)**0.5 / smartphones['DISPLAY_IN_INCHES']

# Function to classify screen resolution based on pixel density
def classify_resolution(pixel_density):
    if pixel_density < 250:
        return 'Standard'
    elif pixel_density >= 250 and pixel_density < 300:
        return 'HD (720p) - 300PPI'
    elif pixel_density >= 300 and pixel_density < 400:
        return 'Full HD (1080p)'
    elif pixel_density >= 400 and pixel_density < 500:
        return 'Quad HD (1440p)'
    elif pixel_density >= 500:
        return 'Ultra HD (4K - 2160p)'
    else:
        return 'Unknown Resolution'

# Applying the function to classify screen resolution for each row
smartphones['Resolution'] = smartphones['Pixel Density'].apply(classify_resolution)

# Inserting the new column 'Resolution' after 'PIXLES_IN_HEIGHT'
column_index = smartphones.columns.get_loc('PIXLES_IN_HEIGHT') + 1
smartphones.insert(column_index, 'Resolution', smartphones.pop('Resolution'))
```

```python
# deleting the remaining unnecessary columns
smartphones.drop(columns=['PIXLES_IN_HEIGHT','PIXLES_IN_WIDTH','Pixel Density',], inplace=True)
```

```python
col = ['#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']
smartphones['Resolution'].value_counts().plot(kind='pie', autopct='%0.1f%%', colors=col,
    wedgeprops=dict(edgecolor='black'))
plt.ylabel('')
plt.xlabel('Smartphones Resolutions')
plt.show()
```
